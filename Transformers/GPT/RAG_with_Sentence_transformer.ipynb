```python
from transformers import GPT2Tokenizer, GPT2LMHeadModel
import torch
from sklearn.metrics.pairwise import cosine_similarity
from sentence_transformers import SentenceTransformer

# Load models
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
model = GPT2LMHeadModel.from_pretrained("gpt2")
embedder = SentenceTransformer("all-MiniLM-L6-v2")

# Knowledge base chunks
kb_chunks = [
    "Thinkster Math is a Knowledge Accelerator and online math tutoring program for K-12.",
    "It combines personalized coaching, AI/technology, and worksheets.",
    "Students take a diagnostic to find strengths and weaknesses, then a learning plan is created.",
    "It is used in 30+ countries and partnered with Teach Away.",
]

# Query
query = "tell me about Thinkster Math?"

# Retrieve most relevant chunk
query_emb = embedder.encode([query])
kb_embs = embedder.encode(kb_chunks)
sims = cosine_similarity(query_emb, kb_embs)[0]
best_chunk = kb_chunks[sims.argmax()]

# Build input
prompt = f"Question: {query}\nContext: {best_chunk}\nAnswer:"
input_ids = tokenizer.encode(prompt, return_tensors="pt")

# Generate
output_ids = model.generate(
    input_ids,
    max_length=120,
    do_sample=True,
    top_k=50,
    top_p=0.95,
    num_return_sequences=1
)

print(tokenizer.decode(output_ids[0], skip_special_tokens=True))
```
